### We look for tech companies in San Francisco

```python
db = client["Ironhack"]
collection = db["companies"]

# Get distinct values from the "category_code" field
distinct_category_codes = collection.distinct("category_code")

# Print the distinct category codes
for category_code in distinct_category_codes:
    print(category_code)
```

```python
def companies_data(category_codes):
    condition_1 = {"$or": [{"category_code": code} for code in category_codes]}
    condition_2 = {"number_of_employees": {"$gte": 77, "$lte": 100}}
    condition_3 = {"funding_rounds": {"$elemMatch": {"raised_amount": {"$gte": 1000000}}}}

    projection = {"_id": 0, "name": 1, "number_of_employees": 1, "category_code": 1, "offices": 1, "funding_rounds.raised_amount": 1}

    result = list(c.find({"$and": [condition_1, condition_2, condition_3]}, projection))

    processed_data = []
    for doc in result:
        if "offices" in doc and len(doc["offices"]) > 0:
            office = doc["offices"][0]
            latitude = office.get("latitude")
            longitude = office.get("longitude")
            country_code = office.get("country_code")
            city = office.get("city")

            if latitude is not None and longitude is not None:
                for funding_round in doc.get("funding_rounds", []):
                    raised_amount = funding_round.get("raised_amount")
                    if pd.notna(raised_amount):  # Filtrar valores no NaN
                        processed_data.append({
                            "name": doc.get("name"),
                            "funding_rounds.raised_amount": raised_amount,
                            "number_of_employees": doc.get("number_of_employees"),
                            "category_code": doc.get("category_code"),
                            "latitude": latitude,
                            "longitude": longitude,
                            "country_code": country_code,
                            "city": city
                        })

    df = pd.DataFrame(processed_data)

    return df

# Lista de las categorías que deseas filtrar
category_codes = ["advertising", "analytics", "biotech", "cleantech", "design", "web", "games_video"]

df = companies_data(category_codes)
df
```

```python
def companies_data2(category_codes):
    condition_1 = {"$or": [{"category_code": code} for code in category_codes]}
    condition_3 = {"funding_rounds": {"$elemMatch": {"raised_amount": {"$gte": 1000000}}}
                  }
    projection = {"_id": 0, "name": 1, "number_of_employees": 1, "category_code": 1, "offices": 1, "funding_rounds.raised_amount": 1}

    result = list(c.find({"$and": [condition_1, condition_3]}, projection))

    processed_data = []
    for doc in result:
        if "offices" in doc and len(doc["offices"]) > 0:
            office = doc["offices"][0]
            latitude = office.get("latitude")
            longitude = office.get("longitude")
            country_code = office.get("country_code")
            city = office.get("city")

            if latitude is not None and longitude is not None and city == "San Francisco":
                for funding_round in doc.get("funding_rounds", []):
                    raised_amount = funding_round.get("raised_amount")
                    if pd.notna(raised_amount):  # Filtrar valores no NaN
                        processed_data.append({
                            "name": doc.get("name"),
                            "funding_rounds.raised_amount": raised_amount,
                            "number_of_employees": doc.get("number_of_employees"),
                            "category_code": doc.get("category_code"),
                            "latitude": latitude,
                            "longitude": longitude,
                            "country_code": country_code,
                            "city": city
                        })

    df = pd.DataFrame(processed_data)

    return df

category_codes = ["advertising", "analytics", "biotech", "cleantech", "design", "web", "games_video"]

df = companies_data2(category_codes)
df
```

### We create the startups map with more than 1M raised in San Fracisco city

```python
json_file_path = r"C:\Users\photo\Desktop\Ironhack\projects\project-III-geolocation\data\startups.json"

with open(json_file_path, 'r') as json_file:
    company_data = json.load(json_file)
```

```python
map_center = [37.790998, -122.394668]
companies_startup = folium.Map(location=map_center, zoom_start=15)

for company in company_data:
    name = company['name']
    latitude = company['latitude']
    longitude = company['longitude']
    popup_text = f"{name}<br>Lat: {latitude}<br>Lon: {longitude}"

    icon_color = {
        'hi5': 'mediumorchid',
        '99designs': 'lightgreen',
        'Company3': 'orange',
        'Company4': 'yellow',
    }.get(name, 'yellow')

    custom_icon_html = f'<div style="background-color: {icon_color}; border-radius: 50%; width: 30px; height: 30px; display: flex; justify-content: center; align-items: center;">{name}</div>'

    folium.Marker(
        location=[latitude, longitude],
        popup=popup_text,
        icon=folium.DivIcon(html=custom_icon_html)
    ).add_to(companies_startup)
    
     # Determinar si esta es la empresa que deseas resaltar en un color diferente
    if name == 'hi5':  # Cambia 'Company1' al nombre de la empresa que deseas resaltar
        icon_color = 'red'  # Cambia el color a tu elección para resaltar esta empresa


folder_path = r"C:\Users\photo\Desktop\Ironhack\projects\project-III-geolocation\geomaps"
file_name = 'companies_startup.html'
full_path = os.path.join(folder_path, file_name)

companies_startup.save(full_path)
companies_startup
```
